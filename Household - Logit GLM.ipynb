```python
#Lift/cumulative gains charts (this just defines functions that prepare tables for evaluation graphs) - it keeps only contracts when we predict Cancellation - best setting is to run this code with small Threshold - so all contracts are in the table

import sklearn.metrics

def calc_cumulative_gains(df: pd.DataFrame, actual_col: str, predicted_col:str, probability_col:str):

    df.sort_values(by=probability_col, ascending=False, inplace=True)

    subset = df[df[predicted_col] == True]

    rows = []
    for group in np.array_split(subset, 20):
        score = sklearn.metrics.accuracy_score(group[actual_col].tolist(),
                                                   group[predicted_col].tolist(),
                                                   normalize=False)

        rows.append({'NumCases': len(group), 'NumCorrectPredictions': score})

    lift = pd.DataFrame(rows)

    #Cumulative Gains Calculation
    lift['RunningCorrect'] = lift['NumCorrectPredictions'].cumsum()
    lift['PercentCorrect'] = lift.apply(
        lambda x: (100 / lift['NumCorrectPredictions'].sum()) * x['RunningCorrect'], axis=1)
    lift['CumulativeCorrectBestCase'] = lift['NumCases'].cumsum()
    lift['PercentCorrectBestCase'] = lift['CumulativeCorrectBestCase'].apply(
        lambda x: 100 if (100 / lift['NumCorrectPredictions'].sum()) * x > 100 else (100 / lift[
            'NumCorrectPredictions'].sum()) * x)
    lift['AvgCase'] = lift['NumCorrectPredictions'].sum() / len(lift)
    lift['CumulativeAvgCase'] = lift['AvgCase'].cumsum()
    lift['PercentAvgCase'] = lift['CumulativeAvgCase'].apply(
        lambda x: (100 / lift['NumCorrectPredictions'].sum()) * x)

    #Lift Chart
    lift['NormalisedPercentAvg'] = 1
    lift['NormalisedPercentWithModel'] = lift['PercentCorrect'] / lift['PercentAvgCase']
    lift['rank']=(lift.index+1)*5


    return lift


def plot_lift_chart(lift: pd.DataFrame):
        plt.figure(figsize=(14,4))
        plt.plot(lift['rank'], lift['NormalisedPercentAvg'], 'r-', label='Normalised \'response rate\' with no model')
        plt.plot(lift['rank'], lift['NormalisedPercentWithModel'], 'b-', label='Normalised \'response rate\' with using model')
        plt.legend()
        plt.grid(color='0.65', linestyle='-', linewidth=0.5)
        plt.xticks(np.arange(5, 105, step=5))
        plt.title('Cumulative Lift Plot')
        plt.xlabel('% of top ranked contracts')
        plt.ylabel('Lift')
        plt.show()

from sklearn.metrics import roc_curve, auc

def plot_gains_chart(df, score1, score2, score3, score4, target):
    df1 = df[[score1,target]].dropna()
    fpr1, tpr1, thresholds1 = roc_curve(df1[target], df1[score1])
    ppr1=(tpr1*df1[target].sum()+fpr1*(df1[target].count()- 
    df[target].sum()))/df1[target].count()

    df2 = df[[score2,target]].dropna()	
    fpr2, tpr2, thresholds2 = roc_curve(df2[target], df2[score2])
    ppr2=(tpr2*df2[target].sum()+fpr2*(df2[target].count()- 
    df2[target].sum()))/df2[target].count()

    df3 = df[[score3,target]].dropna()	
    fpr3, tpr3, thresholds3 = roc_curve(df3[target], df3[score3])
    ppr3=(tpr3*df3[target].sum()+fpr3*(df3[target].count()- 
    df3[target].sum()))/df3[target].count()

    df4 = df[[score4,target]].dropna()	
    fpr4, tpr4, thresholds4 = roc_curve(df4[target], df4[score4])
    ppr4=(tpr4*df4[target].sum()+fpr4*(df4[target].count()- 
    df4[target].sum()))/df4[target].count()
    
    plt.figure(figsize=(18,9))
    plt.plot(ppr1, tpr1, color='blue', label='Stronger Restrictions')
    plt.plot(ppr2, tpr2, color='green', label='Weaker Restrictions')
    plt.plot(ppr3, tpr3, color='orange', label='OverFitted')
    plt.plot(ppr4, tpr4, color='black', label='Tuned')
    plt.legend()
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xticks(np.arange(0,1,step=0.1))
    plt.xlim([0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.grid(b=True, which='both', color='0.65',linestyle='-')
    plt.xlabel('%Population')
    plt.ylabel('%Target')
    plt.title('Cumulative Gains Chart')
```
